```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
import joblib
```

```python
# load data QCM10
QCM10 = 'https://raw.githubusercontent.com/i-wijaya-agriaku/agriaku_20230804_qcm_dataset/master/dataset/QCM10.csv'
QCM10 = pd.read_csv(QCM10, sep=';', header=0)
QCM10.head()
```

```python
# load data QCM12
QCM12 = 'https://raw.githubusercontent.com/i-wijaya-agriaku/agriaku_20230804_qcm_dataset/master/dataset/QCM12.csv'
QCM12 = pd.read_csv(QCM12, sep=';', header=0)
QCM12.head()
```

```python
# load data apartemen
QCM3 = 'https://raw.githubusercontent.com/i-wijaya-agriaku/agriaku_20230804_qcm_dataset/master/dataset/QCM3.csv'
QCM3 = pd.read_csv(QCM3, sep=';', header=0)
QCM3.head()
```

```python
# load data apartemen
QCM6 = 'https://raw.githubusercontent.com/i-wijaya-agriaku/agriaku_20230804_qcm_dataset/master/dataset/QCM6.csv'
QCM6 = pd.read_csv(QCM6, sep=';', header=0)
QCM6.head()
```

```python
# load data apartemen
QCM7 = 'https://raw.githubusercontent.com/i-wijaya-agriaku/agriaku_20230804_qcm_dataset/master/dataset/QCM7.csv'
QCM7 = pd.read_csv(QCM7, sep=';', header=0)
QCM7.head()
```

Menggabungkan semua dataset

```python
data = pd.concat([QCM3,QCM6,QCM7,QCM10,QCM12])
data.head()
```

```python
data.index = [i for i in range(125)]
data.index
```

```python
print("Shape of dataset: ", data.shape)
```

```python
# Cek data null
data.isnull().sum()
```

```python
# Melihat karakteristik data
data.describe()
```

```python
# Melihat korelasi data
corr = data.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(10, 8))
#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```

```python
# Merubah value menjadi satu kolom bernama alcohol

one_Octanol = data[data['1-Octanol']==1]
one_Octanol['Alcohol']=1

one_Propanol = data[data['1-Propanol']==1]
one_Propanol['Alcohol']=2

two_Butanol = data[data['2-Butanol']==1]
two_Butanol['Alcohol']=3

two_propanol = data[data['2-propanol']==1]
two_propanol['Alcohol']=4

one_isobutanol = data[data['1-isobutanol']==1]
one_isobutanol['Alcohol']=5
```

```python
data = pd.concat([one_Octanol,one_Propanol,two_Butanol,two_propanol,one_isobutanol])
data
```

```python
# Menghapus data karena sudah diubah menjadi satu kolom
data = data.drop(['1-Octanol','1-Propanol','2-Butanol','2-propanol','1-isobutanol'],axis=1)
```

```python
data.head()
```

```python
data.shape
```

```python
x= data.iloc[:,:10]
y= data.iloc[:,10:11]
print(x.shape,y.shape)
```

```python
# Membagi dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

MACHINE LEARNING

```python
# Random Forest Clasifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Membuat prediksi
y_pred = model.predict(X_test)
```

```python
y_test= np.array(y_test.to_numpy().T.flat)
submission = pd.DataFrame({'Test':y_test,'Predict':y_pred})
```

```python
# Evaluasi model
report = classification_report(y_test, y_pred)
print(report)
```

```python
# Misalkan y_test adalah nilai target asli dan predict adalah nilai prediksi dari model
confusion = confusion_matrix(y_test, y_pred)

# Membuat heatmap confusion matrix
sns.heatmap(confusion, annot=True, cmap = 'viridis')
plt.show()
```

ANN

```python
# Inisialisasi model MLP
mlp_model = MLPClassifier(hidden_layer_sizes=(50, 10), max_iter=1000, random_state=42, verbose=1)

# Melatih model dengan data latih
mlp_model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred_mlp = mlp_model.predict(X_test)
```

```python
# Menampilkan hasil evaluasi
print("Classification Report for MLP:")
print(classification_report(y_test, y_pred_mlp))
```
